Naming conventions vary a bit, but first brackets always tell the target instrument in the clip.

examples namings from the set:

'207__[voi][dru][cou_fol]2414__3.wav'
    HUMAN VOICE, DRUMS, COUNTRY FOLK

'[cel][cla]0080__1.wav'

````python
import os
````

````python
TRAIN_DIR = "/home/samuel/Programming/ML/datasets/IRMAS/IRMAS-TrainingData"

TEST_DIR1 = "/home/samuel/Programming/ML/datasets/IRMAS/IRMAS-TestingData-Part1/Part1"
TEST_DIR2 = "/home/samuel/Programming/ML/datasets/IRMAS/IRMAS-TestingData-Part2/IRTestingData-Part2"
TEST_DIR3 = "/home/samuel/Programming/ML/datasets/IRMAS/IRMAS-TestingData-Part3/Part3"


test_text_files = [f for f in os.listdir(TEST_DIR3) if f.endswith('.txt')]

instrument_counts = {}

for f_name in test_text_files:
    path = os.path.join(TEST_DIR3, f_name)
    with open(path, 'r') as f:
        count = len(f.readlines())
    
    instrument_counts[count] = instrument_counts.get(count, 0) + 1

print(instrument_counts)
````

````python
train_file_count_per_instr = {
    'cel': 0,
    'cla': 0,
    'flu': 0,
    'gac': 0,
    'gel': 0,
    'org': 0,
    'pia': 0,
    'sax': 0,
    'tru': 0,
    'vio': 0,
    'voi': 0
}

for root, dirs, files in os.walk(TRAIN_DIR):
    folder_name = os.path.basename(root)
    
    if folder_name in train_file_count_per_instr:
        file_count = len(files)
        train_file_count_per_instr[folder_name] = file_count

print(train_file_count_per_instr)
````

````python
import wave

test_wav_files1 = [f for f in os.listdir(TEST_DIR1) if f.endswith('.wav')]
test_wav_files2 = [f for f in os.listdir(TEST_DIR2) if f.endswith('.wav')]
test_wav_files3 = [f for f in os.listdir(TEST_DIR3) if f.endswith('.wav')]


def get_duration_in_s(test_wav_files, directory):
    total_duration_s = 0

    for f in test_wav_files:
        file_path = os.path.join(directory, f)
        with wave.open(file_path, 'rb') as wav_file:
            frames = wav_file.getnframes()
            rate = wav_file.getframerate()
            duration = frames / float(rate)
            total_duration_s += duration

    return total_duration_s

l1 = get_duration_in_s(test_wav_files1, TEST_DIR1)
l2 = get_duration_in_s(test_wav_files2, TEST_DIR2)
l3 = get_duration_in_s(test_wav_files3, TEST_DIR3)

hours = (l1 + l2 + l3) / 60 / 60

print(hours)
````
